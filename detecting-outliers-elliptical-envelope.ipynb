Our data is normally distributed. So we can apply EllipticEnvelope() function to detect outliers.

# Detecting outliers

```python
elpenv = EllipticEnvelope(contamination=0.1, random_state=2)
```

Since the dataset doesn't contain very frequent outliers, and they are sparsely distributed, we set the contamination hyperparameter to a small value of 0.03 (on a scale of 0 to 0.5)

```python
# Returns 1 for Inliers and -1 for outliers
pred = elpenv.fit_predict(X.iloc[:500]) #this fits he model and predicts in a single command works same as fit X and predict X
```

Extracting outliers below

```python
outlier_index = np.where(pred==-1)
outlier_index  
outlier_values = X.iloc[outlier_index]
```

```python
sns.scatterplot(x=X['a1'][:500], y=X['a2'][:500], color = 'b')
sns.scatterplot(x=outlier_values['a1'][:500],y=outlier_values['a2'][:500], color='r')
```

Majority of the outlier in a selected sample of first 500 rows of the shuffled dataset are covered in red.
